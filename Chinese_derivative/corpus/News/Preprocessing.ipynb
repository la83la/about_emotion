```python
import pandas as pd
import numpy as np
import os
```

```python
data_list = os.listdir()
news_list = data_list.copy()

# News list
unwanted = ['.ipynb_checkpoints', 'Preprocessing.ipynb']
news_list = [ele for ele in news_list if ele not in unwanted]
news_list
```

```python
# news
news_combine = []
for dataset in news_list:
    tmp = pd.read_pickle(dataset)
    news_combine.append(tmp)
news_merge = pd.concat(news_combine).reset_index(drop=True)
```
