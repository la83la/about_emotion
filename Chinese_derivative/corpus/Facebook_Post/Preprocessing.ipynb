```python
import pandas as pd
import numpy as np
import os
```

```python
data_list = os.listdir()
fb_list = data_list.copy()


unwanted = ['.ipynb_checkpoints', 'Preprocessing.ipynb']

fb_list = [ele for ele in fb_list if ele not in unwanted]
```

```python
# fb
fb_combine = []
for dataset in fb_list:
    tmp = pd.read_pickle(dataset)
    fb_combine.append(tmp)
fb_merge = pd.concat(fb_combine).reset_index(drop=True)
```
