### EWMA(Exponentially Weighted Moving Average) -> Âπ≥ÊªëÂåñ
ÂÖ¨Âºè : EMAt =  Œ±Price+(1‚àíŒ±)EMAt‚àí1 <br>
ÊÑèÊ∂µ : ÂæûËµ∑ÈªûÈñãÂßãËá≥‰ªäÊâÄÊúâÊ≠∑Âè≤ÂÉπÊ†ºÁöÑ‰∏ÄÂÄãÂä†Ê¨äÂπ≥ÂùáÂÄº

```python
def ewm(arr, span): 
    n0 = arr[0] 
    res = [] 
    res.append(n0) 
    alpha = 2 / (span + 1) 
    
    prev = n0 
    for num in arr[1:]: 
        if not np.isnan(num): 
            new = (alpha * num + (1 - alpha) * prev) 
            res.append(new)         
            prev = new 
        return res 
```
